```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
import seaborn as sns
import urllib.request


diabetes= datasets.load_diabetes()
print(diabetes.DESCR)
```

```
webUrl  = urllib.request.urlopen('https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt')
data = webUrl.read().decode("utf-8")
lines=data.split('\n')
keys=lines[0].split('\t')
print(keys)
data=[]
for i in range(1,len(lines)-1):
  data.append(list(map(float, lines[i].split('\t')))) 
diabetes_data = pd.DataFrame(data, columns=keys)
diabetes_data.isnull().any()
diabetes_data
```

```
correlation_matrix = diabetes_data.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True, linewidths = 0.5)
```

```
sns.set(rc={'figure.figsize':(11,8)})

sns.distplot(diabetes_data['Y'], bins=40)
plt.show()
```

```
X = diabetes_data[['BMI', 'BP','S5']]
Y = diabetes_data['Y']


from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
import statsmodels.api as sm
X_train = sm.add_constant(X_train)
model = sm.OLS(Y_train, X_train).fit()

# Print out the statistics
print(model.summary())
```

```
print(model.params)
ypred=model.params['const'] +model.params['BMI']*X_test['BMI']+model.params['BP']*X_test['BP']+model.params['S5']*X_test['S5']
print(ypred)
plt.scatter(Y_test,ypred)
plt.xlabel('Actual y')
plt.ylabel('Predicted y')
```
